```python
# استيراد المكتبات
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (14, 8)

# تحميل البيانات
df = pd.read_csv(r"C:\Users\SOFT ZONE\Desktop\company_employee_details.csv")

# حذف الأعمدة غير المفيدة
df.drop(columns=['Unnamed: 0'], inplace=True)

# عرض أول صفوف البيانات
print("First 5 rows:")
print(df.head())

# -------------------------------
# 1. Dataset Info
print("\nDataset Info:")
print(df.info())

# -------------------------------
# 2. Statistical Summary
print("\nStatistical Summary:")
print(df.describe())

# -------------------------------
# 3. Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# -------------------------------
# 4. Employment Type Distribution
job_type_counts = df[['full_time', 'part_time', 'contractor']].sum()
print("\nEmployment Type Distribution:")
print(job_type_counts)

# -------------------------------
# 5. Top Companies and Departments
print("\nTop Companies by Employee Count:")
print(df['company'].value_counts().head())

print("\nTop Departments by Employee Count:")
print(df['department'].value_counts().head())

# -------------------------------
# 6. Salary vs Experience Correlation
print("\nCorrelation between salary and experience:")
print(df[['salary', 'prior_years_experience', 'years_in_the_company']].corr())

# -------------------------------
# 7. Plots

plt.figure(figsize=(14, 8))

# 1. Salary Distribution
plt.subplot(2, 2, 1)
sns.histplot(df['salary'], kde=True, bins=30, color='skyblue')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")

# 2. Annual Bonus Distribution
plt.subplot(2, 2, 2)
sns.histplot(df['annual_bonus'], kde=True, bins=30, color='lightgreen')
plt.title("Annual Bonus Distribution")
plt.xlabel("Annual Bonus")
plt.ylabel("Number of Employees")

# 3. Employment Type Pie Chart
plt.subplot(2, 2, 3)
plt.pie(job_type_counts, labels=job_type_counts.index, autopct='%1.1f%%',
        colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title("Employment Type Breakdown")

# 4. Salary by Company
plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='company', y='salary')
plt.title("Salary Distribution by Company")
plt.xlabel("Company")
plt.ylabel("Salary")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
